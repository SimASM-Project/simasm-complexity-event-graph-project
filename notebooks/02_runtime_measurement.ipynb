# Runtime Measurement via SimASM Experimenter

This notebook measures simulation wall-clock runtime for each queueing network model
using the `%%simasm experiment` magic. Each model has its own cell so experiments can
be run, inspected, and re-run independently.

## Experimental Design

| Parameter | Value |
|-----------|-------|
| Replications | 30 |
| Warmup period | 1,000 time units |
| Measurement period | 9,000 time units (1,000 to 10,000) |
| Total run length | 10,000 time units |
| Seed strategy | Incremental from base seed 1 |
| Statistic collected | `departure_count` (observation at end of run) |

## Models

- **Dev set (27 models):** tandem, fork_join, feedback â€” sizes 1, 2, 3, 4, 5, 7, 10, 15, 20
- **OOS set (6 models):** hybrid â€” configs (2,2), (2,3), (3,2), (3,3), (5,2), (5,3)

## Output

Each experiment writes its result JSON to `runtime_results/{model_name}.json`.
The final cells aggregate all results into a single summary table.

## Setup

```python
import simasm
from pathlib import Path

# Create output directory for experiment results
Path('runtime_results').mkdir(exist_ok=True)

print(f'SimASM version: {simasm.__version__}')
print('Output directory: runtime_results/')
```

## Collect Results

Read all per-model JSON result files and aggregate into a single table.

Each JSON file (written by the `output:` block above) contains:
- `replications[].wall_time` — wall-clock seconds per replication (if available)
- `metadata.total_wall_time` — total wall-clock seconds (always present)
- `replications[].statistics.departure_count` — total departures
- `replications[].seed` — random seed used
- `summary` — mean, std, CI for each statistic across replications

> **Note:** If per-replication `wall_time` is not present in the JSON
> (older SimASM output), the mean is computed from `total_wall_time / n`
> and per-replication variance is unavailable.

```python
import json
import numpy as np
from pathlib import Path

RESULTS_DIR = Path('runtime_results')

# Define expected models in order
SIZES = [1, 2, 3, 4, 5, 7, 10, 15, 20]
HYBRID_CONFIGS = [(2,2), (2,3), (2,4), (3,2), (3,3), (3,4), (4,2), (4,3), (4,4)]

model_order = []
for topo in ['tandem', 'fork_join', 'feedback']:
    for n in SIZES:
        model_order.append((f'{topo}_{n}', topo, False))
for n, m in HYBRID_CONFIGS:
    model_order.append((f'hybrid_{n}_{m}', 'hybrid', True))

# Collect
results_data = []
missing = []

for model_name, topology, is_oos in model_order:
    fpath = RESULTS_DIR / f'{model_name}.json'
    if not fpath.exists():
        missing.append(model_name)
        continue

    with open(fpath) as f:
        data = json.load(f)

    # wall_time may be per-replication or only in metadata
    if data['replications'] and 'wall_time' in data['replications'][0]:
        wall_times = [r['wall_time'] for r in data['replications']]
    else:
        # Fall back to total_wall_time / num_replications
        total = data.get('metadata', {}).get('total_wall_time', 0)
        n_reps = len(data['replications'])
        wall_times = [total / n_reps] * n_reps if n_reps > 0 else [0]

    departures = [
        r.get('statistics', {}).get('departure_count', 0)
        for r in data['replications']
    ]

    results_data.append({
        'model_name': model_name,
        'topology': topology,
        'is_held_out': is_oos,
        'wall_time': float(np.mean(wall_times)),
        'wall_time_std': float(np.std(wall_times, ddof=1)),
        'departures': int(np.mean(departures)),
        'runtimes': wall_times,
    })

if missing:
    print(f'Missing results for: {missing}')
print(f'Collected results for {len(results_data)}/{len(model_order)} models')
```

## Summary Table

```python
print(f"{'Model':<20s} {'Topo':<12s} {'Mean(s)':>8s} {'Std(s)':>8s} {'95% CI':>20s} {'Deps':>8s}")
print('-' * 74)

for entry in results_data:
    rts = np.array(entry['runtimes'])
    mean = np.mean(rts)
    n = len(rts)
    # std/CI only meaningful when per-replication wall_time was recorded
    if len(set(rts)) > 1:
        std = np.std(rts, ddof=1)
        ci = 1.96 * std / np.sqrt(n)
        ci_str = f'[{mean-ci:.3f}, {mean+ci:.3f}]'
        std_str = f'{std:.3f}'
    else:
        std_str = '   N/A'
        ci_str = '          N/A'
    print(f"{entry['model_name']:<20s} {entry['topology']:<12s} "
          f"{mean:>8.3f} {std_str:>8s} {ci_str:>20s} {entry['departures']:>8d}")
```
