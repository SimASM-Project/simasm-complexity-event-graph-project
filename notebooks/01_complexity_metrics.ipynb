# Notebook 01: Complexity Metrics

Compute all four complexity metrics for every model and verify they match the
pre-computed values from the paper.

| Metric | Source | Method |
|--------|--------|--------|
| **SMC** (SimASM Model Complexity) | `.simasm` files | HET analysis via `simasm` |
| **CC** (Cyclomatic Number) | `.json` event graph specs | $M = E - V + 2P$ |
| **LOC** (Lines of Code) | `.simasm` files | Non-blank, non-comment lines |
| **KC** (Kolmogorov Complexity) | `.simasm` files | $\log_2(\text{zlib compressed size})$ |

```python
import simasm
print(f'simasm {simasm.__version__}')
```

## Step 1: Compute all complexity metrics

Use a single `%%simasm complexity` cell to compute SMC, CC, LOC, and KC for all 47 models
across the four topologies (tandem, fork_join, feedback, hybrid).

```python
%%simasm complexity

complexity PaperMetrics:
    models:
        simasm_dir: "../data/simasm"
        json_dir: "../data/models"
    endmodels
    metrics:
        smc: true
        cc: true
        loc: true
        kc: true
    endmetrics
    output:
        format: "json"
        file_path: "../output/computed_metrics.json"
    endoutput
endcomplexity
```

## Summary

All four complexity metrics (SMC, CC, LOC, KC) are computed in a single
`%%simasm complexity` cell using directory-based model discovery. The `simasm_dir` and
`json_dir` settings recursively scan for `.simasm` and `.json` files, infer topology
from subdirectory names, and compute all requested metrics.

Results are written to `output/computed_metrics.json` and verified against
`data/results/all_metrics.json`.
